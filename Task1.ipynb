```python
#Scatter plot to analyze relationships between variables.
sns.scatterplot(data=data, x="sepal_length", y="petal_length", hue="species")
```

```python
#Another Scatter plot
sns.scatterplot(data=data, x='petal_length', y='petal_width')
```

```python
#Histogram to examine data distribution.
sns.histplot(data["sepal_length"], bins=20, kde=True)
```

```python
# Box plot to detect outliers and spread of values.
sns.boxplot(data=data, x="species", y="sepal_length")
```
